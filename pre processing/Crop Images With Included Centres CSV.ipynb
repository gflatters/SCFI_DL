```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import skimage
import glob
import os
import imageio
import shutil
```

```python
#remove samples which didn't get past Vitamica checking
#for name in ['/home/ubuntu/data/resistant','/home/ubuntu/data/susceptible']:
for name in ['/home/ubuntu/data/susceptible']:
    for sample in glob.glob(name+'/sample[0-9]*'):
        print(sample)
        #get all cell nums in sample
        allcellnums = [i[-3:] for i in glob.glob(sample+'/cell[0-9]*')]
        
        #read checked csv file to get correct names
        out_files = glob.glob(sample+'/out/*')
        for i in out_files:
            if 'checked' in i:
                checked_cells_csv_name=i 
        print(out_files)    
        #load the cells in the checked spreadsheet
        good_cells = np.genfromtxt(checked_cells_csv_name, delimiter=',',usecols=1,skip_header=1,dtype='unicode')
        #shave off the paths to just the final cell numbers
        goodcellsnums = [i[-3:] for i in good_cells]
        #find nums which are not in the good cell nums
        badcellnums = list(set(allcellnums).difference(set(goodcellsnums)))
        #delete bad cells
        print(badcellnums)
        for cell in glob.glob(sample+'/cell[0-9]*'):
            if cell[-3:] in badcellnums:
                shutil.rmtree(cell)
```

```python
#crop good cells to required size

#, '/home/ubuntu/data/susceptible'
#,'/home/ubuntu/data/resistant'
#
for name in ['/home/ubuntu/data/cephdata/resistant', '/home/ubuntu/data/cephdata/susceptible']:
#for name in ['/home/ubuntu/data/control']: 
    print(name)
    for path in glob.glob(name+'/sample[0-9]*/cell[0-9]*'):
        createpathname = '/cropped_video100_140' #create a new path for our cropped images
        if os.path.exists(path+createpathname):
            shutil.rmtree(path+createpathname)
        os.mkdir(path+createpathname) #make the directory
        number = int(path[-3:]) #pull out the correct cell number from directory name
        
        path_csv = glob.glob(path+'/*.csv')[0] #make some more paths
        path_videos = path+'/video%03d/'%(number)
        path_save = path+createpathname+'/'
        
        center = np.genfromtxt(path_csv, delimiter=',',usecols=[0,1],skip_header=1,max_rows=1)#find the centre from centres csv
        print(center)
        centerint = center.astype('int') 
        if (centerint<71).any() or (centerint>229).any():
            print('bad centre at',path)
        else: 
            for i,k in enumerate(glob.glob(path_videos+'video*.TIF')):#walk through the videos
                image = plt.imread(k) #read video
                image_crop = image[centerint[1]-70:centerint[1]+70,centerint[0]-50:centerint[0]+50]#crop video
                imageio.imwrite(path_save+'croppedvideo%03d.tiff'%i, image_crop) #save
            print(path)
```
