```python
import pickle
result_all = pickle.load(open('/home/ubuntu/data/code/2dCNN/2d_k5_cip_trim_final.p','rb'))
train_preds = result_all[0]
kfold_val_acc = result_all[1]
kfold_acc = result_all[2]
kfold_val_loss = result_all[3]
#print((list(train_preds)))

preds=[]
[preds.extend([(keys,values)]) for keys,values in train_preds.items()]
preds
```

```python
#seperate predictions into drug type and class
#add extra [0] for 3D
cipro_preds = []
cipro_preds_res = []
cipro_preds_sus = []
trim_preds = []
trim_preds_res = []
trim_preds_sus = []

for path_preds in preds:
    print(path_preds[0])
    if "trimdata" in path_preds[0]:
        trim_preds.append(path_preds)
        if "resistant" in path_preds[0]:
            trim_preds_res.append(path_preds)
        if "susceptible" in path_preds[0]:
            trim_preds_sus.append(path_preds)

    else:
        cipro_preds.append(path_preds)
        if "resistant" in path_preds[0]:
            cipro_preds_res.append(path_preds)
        if "susceptible" in path_preds[0]:
            cipro_preds_sus.append(path_preds)
            
#print(path_preds[0])
print(len(cipro_preds))           
print(len(cipro_preds_res))
print(len(cipro_preds_sus))
print(len(trim_preds))
print(len(trim_preds_res))
print(len(trim_preds_sus))
```

```python
# in prediction array, 1st value is prob(res), 2nd value is prob(sus)
# finding accuracy of sus and res for cipro
cipro_res_count = 0
cipro_sus_count = 0
for data in cipro_preds_res:
    if data[1][0] > 0.5:
        cipro_res_count += 1
for data in cipro_preds_sus:
    if data[1][1] > 0.5:
        cipro_sus_count += 1
print('cipro res acc =', cipro_res_count/len(cipro_preds_res))
print('cipro sus acc =', cipro_sus_count/len(cipro_preds_sus))
```

```python
# in prediction array, 1st value is prob(res), 2nd value is prob(sus)
# want to see accuracy of sus and res by net
trim_res_count = 0
trim_sus_count = 0
for data in trim_preds_res:
    if data[1][0] > 0.5:
        trim_res_count += 1
for data in trim_preds_sus:
    if data[1][1] > 0.5:
        trim_sus_count += 1
print('trim res acc =', trim_res_count/len(trim_preds_res))
print('trim sus acc =', trim_sus_count/len(trim_preds_sus))
```
